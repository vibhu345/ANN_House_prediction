# IMPORTING PACKAGES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split

# deep learning libraries
from keras .models import Sequential
from keras.layers import Dropout,Dense
from keras.callbacks import EarlyStopping
```

```python
# loading dataset
df=pd.read_csv("housing.csv")
df.head(5)
```

```python
# preprocessing
df.shape # since we have large number of rows ie we will use deep learning, we usually use deep learning when data volume is large
df.info()
```

```python
df.ocean_proximity.value_counts()
```

```python
df.isnull().sum() # only total_bedrooms column has 207 null values since our dataset is very large so we can remove these values
df.dropna(inplace=True)
df.duplicated().sum() # no duplicated values
```

```python
# ENCODING
# ocean_proximity has categorical values so we will encode it . we can use inbuilt functions but we will try it doing manually here,just for the sake of exploration
df.ocean_proximity =df.ocean_proximity .map({"<1H OCEAN":0,"INLAND":1,"NEAR OCEAN":2,"NEAR BAY":3,"ISLAND":4})
df.head()
df.ocean_proximity.value_counts()
```

```python
#train test split
x=df.drop("median_house_value",axis=1)
y=df.median_house_value
# Converting x and y into numpy array
x=np.array(x)
y=np.array(y)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
# scalling
scaling=MinMaxScaler()
x_train=scaling.fit_transform(x_train)
x_test=scaling.transform(x_test)
x_train
```

```python
# Model training and architecture
# we will make one input layer ,one output layer and 2 hidden layer
model=Sequential([
    # input layer
    Dense( 1000, #no of neurons
          input_shape=[x_train.shape[1]], # no of columns
         ),
    # hidden layer 1
    Dense(500,activation="relu"),
    # hidden layer 2
    Dense(250,activation="relu"),
    # output layer
    Dense(1,activation="linear") # agar classification problem hoti toh sigmoid activation function use karte

])
model.summary()
```

```python
# compilation of model
model.compile(optimizer="rmsprop",loss="mse",metrics=["mae"])
es=EarlyStopping(monitor="val_loss",mode="min",patience=30,restore_best_weights=True)
history=model.fit(x_train,y_train,validation_data=(x_test,y_test),callbacks=[es],epochs=50,batch_size=50,verbose=1) #epochs=10 means yeh model training data pe 10 times train hoga

#patience = 50 means This callback will stop the training when there is no improvement in the loss for 50(or whatever the value patience has) consecutive epochs.
```

```python
y_pred=model.predict(x_test)
from sklearn.metrics import r2_score,mean_squared_error,mean_squared_log_error,mean_absolute_error
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(mean_squared_log_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
# plotting training loss and validation loss
history.history
training_loss=history.history["loss"]
validation_loss=history.history["val_loss"]
epochs=range(1,len(training_loss)+1)
# plot the training and validation loss
plt.plot(epochs,training_loss,"b",label="Loss during Training")
plt.plot(epochs,validation_loss,"r",label="Loss during validation")

# label the plot
plt.title("Training and validation loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()

# show the plot
plt.show()
```

```python
def pred(longitude,latitude,housing_median_age,total_rooms,total_bedrooms,population,households,median_income,ocean_proximity):
    features=np.array([longitude,latitude,housing_median_age,total_rooms,total_bedrooms,population,households,median_income,ocean_proximity])
    print(features)
    scaled_features=scaling.fit_transform([features])
    result=model.predict(scaled_features).reshape(1,-1)
    print(result)
    return result[0]
```

```python
longitude=-122.2300
latitude=37.8800
housing_median_age=41.0000
total_rooms=880.0000
total_bedrooms=129.0000
population=322.0000
households=126.0000
median_income=8.3252
ocean_proximity=3.0000
price=pred(longitude,latitude,housing_median_age,total_rooms,total_bedrooms,population,households,median_income,ocean_proximity)
print(price)
```
